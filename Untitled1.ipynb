```python
print('This programme helps you generate a shopping list and estimate amount you need to buy all items')
x=[]
y=[]
d=0
e=[]
i= True
while i is True:
    z=input('Do you want to add an item Y/N')
    if z == 'y' or z == 'Y':
        x.append(input('enter the item name'))
        y.append(int(input('enter the price of item mentioned')))
        i=True
    else:
        i=False
a=dict(zip(x,y))
for i in range(0,len(a)):
    b=input('item name')
    c=input('quantity of the item')
    d+=a[b]*int(c)
    e.append(b+'   x'+c+' = '+str(a[b]*int(c)))
print(d,'is estimated price according to you and list of items with quantity needed is')
for i in e:
    print(i)
```
